##### 11. Which are the top 5 selling platforms (games global sales) of all time? Plot your results in a graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('csv/vgsales.csv')
```

```python
top_5_selling_platforms = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(5).sort_index()
top_5_selling_platforms
```

```python
top_5_selling_platforms.plot(kind='bar')
plt.show()
```

##### 12. Which are the top three publishers in EU game sales? Use a pie plot to show their contribution. How is it compared with their global sales? Use plots to show your results.

```python
top_3_pub_in_eu_sales = df.groupby('Publisher')['EU_Sales'].sum().sort_values(ascending=False).head(3).sort_index()
top_3_pub_in_eu_sales
```

```python
top_3_pub_in_eu_sales_global_perf = df[df['Publisher'].isin(top_3_pub_in_eu_sales.index)].groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(3).sort_index()
top_3_pub_in_eu_sales_global_perf
```

```python
top_3_pub_in_eu_sales.plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
top_3_pub_in_eu_sales_global_perf.plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

##### 13. Use map and the following dictionary to replace the genres.

```python
dic = {
1:'Sports',
2:'Platform',
3:'Racing',
4:'Role-Playing',
5:'Puzzle',
6:'Misc',
7:'Shooter',
8:'Simulation',
9:'Action',
10:'Fighting',
11:'Adventure',
12:'Strategy'}
```

```python
df['Genre'] = df['Genre'].map(dic)
df.head()
```

##### 14. Which action game has the highest global sales?

```python
highest_selling_action_game = df[df['Genre'] == 'Action'].groupby('Name').sum(numeric_only=True).sort_values('Global_Sales', ascending=False).head(1)
highest_selling_action_game['Global_Sales']
```

##### 15. Which are the top 5 most mentioned genres of all time? Is this different in 2014, 2015, 2016?

```python
df['Genre'].value_counts().sort_values(ascending=False).head(5).plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
df[df['Year'] == 2014]['Genre'].value_counts().sort_values(ascending=False).head(5).plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
df[df['Year'] == 2015]['Genre'].value_counts().sort_values(ascending=False).head(5).plot(kind='pie', autopct='%1.1f%%')
plt.show()
```
